```python
import numpy as np
import matplotlib.pyplot as plt

r = 1.0
a = np.linspace(0, 2 * np.pi, 360, endpoint=False)
p = [(x, y) for x, y in zip(r * np.cos(a), r * np.sin(a))]
x, y = zip(*p)
plt.figure(figsize=(6, 6))
plt.plot(x, y, "b-")
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect("equal", adjustable="box")
```
